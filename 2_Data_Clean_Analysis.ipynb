**Data Cleaning**

```python
train = train.drop(['Ticket', 'Name', 'Cabin'], axis = 1)
```

```python
train
```

```python
test = test.drop(['Ticket', 'Name', 'Cabin'], axis = 1)
```

```python
test
```

```python
train['PassengerId'].isnull().sum()
train['Pclass'].isnull().sum()
train['Sex'].isnull().sum()
train['Age'].isnull().sum()
train['SibSp'].isnull().sum()
train['Parch'].isnull().sum()
train['Fare'].isnull().sum()
train['Embarked'].isnull().sum()
```

```python
train['Age']=train['Age'].fillna(train['Age'].median())
train['Age'].isnull().sum()
```

```python
test['Age']=test['Age'].fillna(test['Age'].median())
test['Age'].isnull().sum()
```

```python
train['Embarked']=train['Embarked'].fillna(method='pad')
train['Embarked'].isnull().sum()
```

```python
test['Embarked']=test['Embarked'].fillna(method='pad')
test['Embarked'].isnull().sum()
```

```python
test['Fare']=test['Fare'].fillna(test['Fare'].mean())
test['Fare'].isnull().sum()
```

```python
column_train=['Age', 'Pclass', 'SibSp', 'Parch', 'Embarked', 'Fare', 'Sex']
X = train[column_train] #Training
Y = train['Survived'] #Target

X
```

```python
# Converting text into Numbers using Key:Value
d = {'male':0, 'female':1}
X['Sex']=X['Sex'].apply(lambda x:d[x])

test['Sex']=test['Sex'].apply(lambda x:d[x])

X['Sex'].head()

test['Sex'].head()
```

```python
e = {'C':0, 'Q':1, 'S':2}
X['Embarked'] = X['Embarked'].apply(lambda x:e[x])
X['Embarked'].head()

test['Embarked'] = test['Embarked'].apply(lambda x:e[x])
test['Embarked'].head()
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X,Y)
```

```python
column_test=['Age', 'Pclass', 'SibSp', 'Parch', 'Embarked', 'Fare', 'Sex']
X_test = test[column_test]
test.isnull().sum()
```

```python
predicted = model.predict(X_test)
predicted
```

```python
table = pd.DataFrame({'PassengerId': test['PassengerId'], 'Survival_pred':predicted})
table
```
